```python
# Scaling the dummies by their responses
for i in range(len(ords)):
    action=basement[ords[i]].unique()
    response=[basement[0:1459].loc[basement[ords[i]]==x,'SalePrice'].mean() for x in basement[ords[i]].unique()]
    distx=max(action)-min(action)
    disty=max(response)-min(response)
    progress=[(x-min(response))/disty for x in response]
    basement[ords[i]]=basement[ords[i]].replace(dict(zip(action,progress)))
```
